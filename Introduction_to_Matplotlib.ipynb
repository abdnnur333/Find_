# Introduction to **Matplotlib**

### **IMPORTS**

```python
import matplotlib.pyplot as plt   # plotting library
import numpy as np      # numerical library for arrays
```

### **Simple Plot**

If a plot function receives only one list, it is assigned to the `y-axis` by default, while the`x-axis` values are automatically generated as consecutive integers beginning at zero (starting from 0, 1, 2 & so on).


```python
plt.plot([1, 2, 3, 4])
plt.show()
```

## **Simple Plot with x-axis and y-axis**
### A. Cartesian Coordinates (Normal X-Y Plot)

*   Each pair `(x[i], y[i])` is a point on the graph.
*   `plt.plot()` connects the points with a line.
*   `plt.show()` tells Colab to display the result.

```python
plt.plot([12,22,32,42],[10,20,30,40])     # draws a line connecting the points
#plt.plot([12,22,32,42],[10,20,30,40],color ='red', linestyle='--', marker='o')
plt.show()      # displays the graph
```

### B. Polar Coordinates

*   Polar coordinates use an angle `(theta)`, $\theta$ and distance from center `(r)`.
*   `plt.polar()` automatically switches axes to polar mode.
*   Useful for circular shapes and periodic functions.

```python
import numpy as np

theta = [0, 2*np.pi/5, 4*np.pi/5, 6*np.pi/5, 8*np.pi/5, 0]
r = [1, 0.4, 1, 0.4, 1, 1]

# Create polar plot
plt.polar(theta, r)
plt.show()      # displays the graph
```

## **Axes Label and Title**

```python
plt.plot([1,2,3,4],[10,20,30,40])
plt.xlabel('x-axis')      # name for horizontal axis
plt.ylabel('y-axis')      # name for vertical axis
plt.title("Simple Line Plot")# name for entire graph
plt.grid(True)             # Display grid
plt.show()
```

## Plotting with NumPy
NumPy arrays allow vectorized operations, which are faster than Python lists.

```python
x = np.array([1,2,3,4])
y = np.array([10,20,30,40])
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Legend

*   `label="..."` names each plotted line.
*   `plt.legend()` makes those names appear on the graph.
*   This is essential when showing more than one function.


```python
x = np.array([-3., -2., -1., 0., 1., 3.])
y = np.array([-80., -13., 6., 1., 5., 16.])

p = np.array([1,2,3,4])
q = np.array([10,20,30,40])

plt.plot(x,y,'r-')
plt.plot(p,q,'g*')
plt.legend(['red line', 'green star'], loc = 'lower right')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Plotting Multiple Lines and Adding a Legend


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color="blue")  # Line 1
plt.plot(x, y2, label="cos(x)", color="red")   # Line 2
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sine and Cosine Waves")
plt.legend()  # Show legend
plt.show()
```

### Using subplot()

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # 2 rows, 1 column, plot 1
plt.plot(x, y1, color='blue')
plt.title("Sine Wave")

plt.subplot(2, 1, 2)  # 2 rows, 1 column, plot 2
plt.plot(x, y2, color='red')
plt.title("Cosine Wave")

plt.tight_layout()  # Adjusts spacing
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.sqrt(x)

# 2. Initialize a 2-row, 2-column grid canvas
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# 3. Plot Top-Left Graph (Row 0, Column 0)
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].grid(True)

# 4. Plot Top-Right Graph (Row 0, Column 1)
axes[0, 1].plot(x, y2, color='orange')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].grid(True)

# 5. Plot Bottom-Left Graph (Row 1, Column 0)
axes[1, 0].plot(x, y3, color='green')
axes[1, 0].set_title('Quadratic Function')
axes[1, 0].grid(True)

# 6. Plot Bottom-Right Graph (Row 1, Column 1)
axes[1, 1].plot(x, y4, color='red')
axes[1, 1].set_title('Square Root Function')
axes[1, 1].grid(True)

# 7. Add an overall canvas title and auto-adjust spacing
fig.suptitle('2x2 Grid Layout Example', fontsize=16)
plt.tight_layout()

# 8. Render the final figure window
plt.show()
```

### Horizontal and Vertical lines

```python
fig, ax = plt.subplots()
ax.axhline(y=0.3, color='k', label='y')
ax.axvline(x=2, color='b', label='x')
ax.legend()
```
